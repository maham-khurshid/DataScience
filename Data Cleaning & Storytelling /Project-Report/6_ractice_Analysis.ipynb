## Cacluate the Discounts

```python
Products_orderlines = pd.merge(ol_df, products_df, on='sku',how='inner')
```

```python
Products_orderlines.head(2)
```

```python
pd.set_option('display.max_rows', 1000)
pd.set_option('display.min_rows', 1000)
```

```python
 Products_orderlines.loc[:,['id','id_order','name','desc','product_quantity','sku','date','price','unit_price']].head(2)
```

```python
 Products_orderlines['Discounts']= Products_orderlines['price']- Products_orderlines['unit_price']
```

```python
Products_orderlines.head(5)
```

```python
Products_orderlines['Discounts'].describe()
#Interpretation: around 50 % of the data the discounts is between 5.55 and 31.86 euros 
```

```python

Products_orderlines['Discounts_Percentage'] = ((Products_orderlines['Discounts'])/(Products_orderlines['unit_price'])*100).round(2)
```

```python
Products_orderlines.head(5)
```

```python
Products_orderlines.info(5)
```

```python
ol_product_o= pd.merge(Products_orderlines, orders_df, left_on='id_order', right_on='order_id', how='inner')
ol_product_o.head(2)
```

```python
ol_product_o.info()
```

```python
ol_product_o['created_date']= pd.to_datetime(ol_product_o['created_date'])
```

```python
ol_product_o["month_name"] = ol_product_o["created_date"].dt.month_name()
ol_product_o["year"] = ol_product_o["created_date"].dt.year
ol_product_o["day"] = ol_product_o["created_date"].dt.day
```

## Calculate the Revenue from the orderline table 

```python
ol_product_o.head(5)
```

```python
ol_product_o.loc[:,['id_order','name','desc','sku','date','price','unit_price','created_date','Discounts','Discounts_Percentage','product_quantity','unit_price']].head(2)
```

```python
ol_product_o['Revenue'] = ol_product_o['product_quantity'] * ol_product_o['unit_price']
# Check the Calculation step1
#check step 2- group by the order_id and summarising by the sum of the revenue 
```

```python
ol_product_o.head(2)
```

```python
Grouped_order_id =  ol_product_o.groupby(['id_order'],as_index=False).agg({'Revenue':'sum','unit_price':'sum','product_quantity':'sum'}).head(5)
```

```python
Grouped_order_id #Check this is the Correct results as on the NOTEBOOK-Workshop 
#Hint you know why u have the same unit_price and revenue because the product quantity sold is 1:)
```

```python
Grouped_year_month =  ol_product_o.groupby(['year','month_name']).agg({'Revenue':'sum','Discounts_Percentage':'mean'}).round(2)
```
