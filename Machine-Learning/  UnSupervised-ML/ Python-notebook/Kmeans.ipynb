```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import set_config

set_config(transform_output='pandas')
```

```python
url = "https://drive.google.com/file/d/16RXRK8V8cKf1YvG8pz-xN-o9t7SotEhD/view?usp=drive_link"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
songs = pd.read_csv(path)
songs.head()
```

```python
songs = songs.set_index(['name','artist'])
songs = songs.drop(['id','html','type','key','mode','time_signature','duration_ms'],axis=1)
songs.info()
```

## 1. Scaling

```python
from sklearn.preprocessing import RobustScaler

my_scaler = RobustScaler()
my_scaler.fit(songs)
```

```python
songs_scaled = my_scaler.transform(songs)
```

### 2. Clustering

```python
from sklearn.cluster import KMeans

my_kmeans = KMeans(n_clusters=16, random_state=356)
my_kmeans.fit(songs_scaled)
songs['cluster'] = my_kmeans.predict(songs_scaled)
```
